```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("여가활동통계.xlsx", header=None)
```

```python
data.iloc[57] = data.iloc[57].replace(method='ffill')
```

```python
data1 = data.iloc[57:67, :12].drop(1,axis=1).reset_index(drop=True)
data1
```

```python
data1.columns = data1.iloc[0].str.cat(data1.iloc[1], sep=' ')
```

```python
data1
```

```python
data1 = data1.drop([0,1], axis=0).reset_index(drop=True)
```

```python
col = data1.columns.tolist()
col[0] = "분류"
data1.columns = col
```

```python
data1.set_index("분류", inplace=True)
```

```python
data1
```

```python
data2=data1.T
data2
```

```python
data2["합"] = data2.apply(np.sum, axis=1)
data2
```

```python
data3 = data2.iloc[:,:8].div(data2.iloc[:,8], axis=0)
```

```python
data3
```

```python
people_ratio = [159,179,122,143,93,114,45,71,21,53]
data4 = pd.DataFrame()
for i, ratio in enumerate(people_ratio):
    data4 = data4.append([data3.iloc[i]]*ratio)
```

```python
data4
```

```python
data4.groupby(level=0).mean()
```

```python
np.set_printoptions(suppress = False)
data5 = pd.DataFrame()
for index in data4.index.unique():
    temp = data4.loc[index,:]
    for column in data4.columns:
        normal_random = np.random.normal(temp.loc[index,column].iloc[0], 1, len(temp))
        temp.loc[:,column] = normal_random
    data5 = data5.append(temp)
```

```python
data5.groupby(level=0).mean()
```

```python
data5[~(data5>0).all(1)]
```
